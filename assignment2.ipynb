```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
patient1 = [76,126,38.0]
patient2 =[74,120,38.0]
patient3 = [72,118,37.5]
patient4 = [78,136,37.0]
```

```python
patients = [patient1, patient2, patient3, patient4]
data = np.array(patients)
print(patients)
```

```python
num_points = data.shape[0]

farthest_patient_index = -1
max_distance = 0

for i in range(num_points):
    distances = [np.linalg.norm(data[i] - data[j]) for j in range(num_points) if i != j]
    closest_distance = max(distances)

    if closest_distance > max_distance:
        max_distance = closest_distance
        farthest_patient_index = i

print(f"The patient farthest from the rest is Patient {farthest_patient_index + 1} with a distance of {max_distance:.2f}.")
```

```python
num_patients = data.shape[0]
distances = np.zeros((num_patients, num_patients))

for i in range(num_patients):
    distances[i] = [np.linalg.norm(data[i] - data[j]) for j in range(num_patients)]
    distances[i, i] = np.inf

min_distance_index = np.unravel_index(np.argmin(distances), distances.shape)
patient1_index, patient2_index = min_distance_index

min_distance = distances[patient1_index, patient2_index]

print(f"Patient {patient1_index + 1} and Patient {patient2_index + 1} are nearest.")
print("Distance:", min_distance)
```

```python
num_patients = data.shape[0]
distances = np.linalg.norm(data[:, np.newaxis] - data, axis=2)

new_dummy_patient = np.array([75, 130, 37.8])

new_patient_distances = np.linalg.norm(data - new_dummy_patient, axis=1)

closest_patient_index = np.argmin(new_patient_distances)

print(f"The closest patient to the new dummy patient is Patient {closest_patient_index + 1}")
print("Distance:", new_patient_distances[closest_patient_index])
```
